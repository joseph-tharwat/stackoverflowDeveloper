```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH = r"..\data\raw\survey_results_public.csv"
EXPLOR_PATH = r"..\data\processed\preprocessed_1.csv.pkl"

REPLACE_DECT = {
    "YearsCode":{
        "More than 50 years":51, #maybe would be more but no information to set 
        "Less than 1 year": 0,
        np.nan: 0},
    "YearsCodePro":{
        "More than 50 years":51, #maybe would be more but no information to set 
        "Less than 1 year": 0,
        np.nan: 0
    }
                                }

pd.options.display.max_columns=None
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
def split_data(data, separator=';'):
    modified_data = []
    for entity in data:
        if(pd.isna(entity) == True):
            empty_list = []
            modified_data.append(empty_list)
        else:
            modified_data.append(entity.split(separator))
    
    return modified_data

split_data(df["DevType"])
```

```python
df2 = df.copy()
```

## Split the answer ;

```python
object_columns = df.select_dtypes(include=object).columns.to_list()
for col in object_columns:
    df2[col] = split_data(df[col])
```

## Replace the string values with numbers in some columns

```python
for col, dect in REPLACE_DECT.items():
    # df2[col] = df[col].replace(dect).astype(np.float64)
    df2[col] = df[col].replace(dect)
    df2[col] = df2[col].apply(lambda s: float(s))
```

## check if the splitting is correct
